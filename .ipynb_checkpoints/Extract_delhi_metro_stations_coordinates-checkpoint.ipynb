### Web Scraping using Beautiful Soup

The metro stations in Delhi are classified as various lines denoted by colors (Blue line, Red line, Yellow line etc.). In order to understand the popularity of restaurants it will be beneficial to have a look at the public transport of Delhi. Since the advent of metro stations has made it much convenient to travel within Delhi, making the proximity of a restaurant to a metro station a likely parameter of popularity. It is also it is highly likely that prices of the restaurants would also depend upon the proximity to a metro station. I will first extract the stations according to the color code of the lines. This would make it easier to visualize.


Beautiful Soup 4 is a python library that can be used to scrape websites and we will use this to obtain the list of names from wikipedia.

```python
from bs4 import BeautifulSoup
import urllib.request as urllib
from geopandas import GeoDataFrame
from shapely.geometry import Point, LineString
```

```python
#define function for extracting the names of metro stations
def extract_names(url, table_no):
    header = {'User-Agent': 'Mozilla/5.0'} #Needed to prevent 403 error on Wikipedia
    req = urllib.Request(url ,headers=header)
    page = urllib.urlopen(req)
    soup = BeautifulSoup(page, "lxml")
    table = soup.findAll("table", { "class" : "wikitable"})[table_no]
    station_names = []

    for row in table.findAll("tr"):

        cell = row.findAll('td')
        try:
            station_names.append(cell[1].find(text = True))
            #station_name_Hindi.append(cell[2].find(text = True))
            #station_line.append(cell[3].find('a').contents[0])
            #print(cell[2].find(text=True))
        except IndexError:
            pass
    return station_names
```

```python
#extract station location using the google API
def extract_locations(station_list):
    locations = []
    for name in station_list:
        location = gmaps.geocode(name+' Metro Station, Delhi, India')
        try:
            if len(location)==2:
                latlon = location[1]['geometry']['location']
            else:
                latlon = location[0]['geometry']['location']
            geolatlon = Point(latlon['lng'], latlon['lat'])
            locations.append(geolatlon)
            #print(location)
        except IndexError:
            print(name)
    return LineString(locations)
```

#### Extracting stations of the Blue Line of Delhi Metro

```python
wiki = "https://en.wikipedia.org/wiki/Blue_Line_(Delhi_Metro)"
station_name_blue_line = extract_names(wiki, 1)
station_name_blue_line = station_name_blue_line[6:]
```

```python
blue_line = extract_locations(station_name_blue_line)
```

```python
crs = {'init': 'epsg:4326'}
blue = GeoDataFrame(crs = crs, geometry=[blue_line])
```

### Overlay of metro lines on map of Delhi

```python
import geopandas as gpd
delhi = gpd.read_file('delhi_map_full.geojson')
```

```python
blue = blue.to_crs(delhi.crs)
yellow = yellow.to_crs(delhi.crs)
red = red.to_crs(delhi.crs)
green = green.to_crs(delhi.crs)
green_br = green_br.to_crs(delhi.crs)
violet = violet.to_crs(delhi.crs)
magenta = magenta.to_crs(delhi.crs)
orange = orange.to_crs(delhi.crs)
```

```python
base = delhi.plot(color = 'white', edgecolor = 'black', figsize = (20, 20))
blue.plot(ax = base, color = 'Blue', figsize = (12, 9))
yellow.plot(ax = base, color = 'Orange', figsize = (12, 9))
red.plot(ax = base, color = 'Red', figsize = (12, 9) )
green.plot(ax = base, color = 'Green', figsize = (12,9))
green_br.plot(ax = base, color = 'Green', figsize = (12,9))
violet.plot(ax = base, color = 'Violet', figsize = (12, 9))
magenta.plot(ax = base, color = 'Magenta', figsize = (12, 9))
orange.plot(ax = base, color = 'Red', figsize = (12, 9))
```
